### EOS 

```python
%%time
#fittype bm or vinet
fittype = 'birchmurnaghan'

# Create murnaghan job for fitting
pr = Project('dummy')
murn = pr.create_job(pr.job_type.Murnaghan, 'birchmurn')

# eos_plot_data holds a tuple for each pot and basis combination.
# Tuple => (volume, energy, states, errors)
eos_plot_data = []
eos_df_rows = []
eos_paths = []

# Obtain eos data
for pr in [Database1]:
    for structure in pr.list_groups():
        pr_s = pr[structure]
        for potential in pr_s.list_groups():
            pr_sp = pr_s[potential]
            
            #volume and energy and path for eos fits
            vol_eos, en_eos = [], []
            eos_paths.append(pr_sp.path.replace('/cmmc/u/caras/',''))
            for strain in pr_sp.list_groups():
                pr_sps = pr_sp[strain]
                
                # YOU HAVE ENERGY,PHONON,TILD JOBS HERE.
                energyjob = pr_sps['energyjob']
                at_nu = energyjob.structure.get_number_of_atoms()
                vol_eos.append(energyjob['output/generic/volume'][-1]/at_nu)
                en_eos.append(energyjob['output/generic/energy_tot'][-1]/at_nu)
                
            states, errors = murn._fit_leastsq(volume_lst=vol_eos,
                                               energy_lst=en_eos,
                                               fittype=fittype)
            eos_plot_data.append((vol_eos, en_eos, states, errors))
            eos_df_rows.append(np.concatenate((states,errors))) 

# Plot EOS df
eos_df = pd.DataFrame(eos_df_rows,
                      columns=[
                          'e0 (perAtom)', 'b0', 'bP', 'v0 (perAtom)', 
                          'e0_err', 'b0_err', 'bP_err', 'v0_err'
                          ])
eos_df.insert(loc=0, column='paths', value=eos_paths)
eos_df.insert(loc=1, column='fittype', value=fittype)
```

### Phonon bandstructure and force constants

#### Check band structures of different crys. structures where there is no strain.
#### Total plot number = structure x potential x 2

```python
%%time
from mylib import get_fc_from_pos_dyn_matrices

force_constants_posdef = []
force_constants_normal = []

# Rows->Volume, Columns->Temperature for surface array
vol_arrays_qha = []
surface_arrays_qha = []
        
phonon_paths = []

temp_array = np.around(np.linspace(0,1600,num=321))
# Obtain phonon data
for pr in [Database1]:
    for structure in pr.list_groups():
        pr_s = pr[structure]
        for potential in pr_s.list_groups():
            pr_sp = pr_s[potential]     
            
            vol_array_normal = np.zeros(len(pr_sp.list_groups()))
            surface_array_normal = np.zeros((len(pr_sp.list_groups()), len(temp_array)))

            vol_array_posdef = np.zeros(len(pr_sp.list_groups()))
            surface_array_posdef = np.zeros((len(pr_sp.list_groups()), len(temp_array)))
            
            for c,strain in enumerate(pr_sp.list_groups()):
                pr_sps = pr_sp[strain]
                # YOU HAVE ENERGY,PHONON,TILD JOBS HERE.
                
                phonon_path = pr_sps.path.replace('/cmmc/u/caras/','')
                
                energyjob = pr_sps['energyjob']
                at_nu = energyjob.structure.get_number_of_atoms()
                U_per_atom = energyjob.output.energy_tot[-1]/at_nu
                vol_per_atom = energyjob.structure.get_volume()/at_nu
                
                phononjob = pr_sps['phononjob']
                
                phonopy_normal = phononjob.phonopy
                phonopy_normal.positivedefdyn = False
                phonopy_normal._set_dynamical_matrix()
                
                phonopy_posdef = deepcopy(phonopy_normal)
                phonopy_posdef.positivedefdyn = True
                phonopy_posdef._set_dynamical_matrix()

                at_nu_prim = phonopy_normal.primitive.get_number_of_atoms()
 
                
               
                phonopy_normal.auto_band_structure(plot=False)
                phonopy_normal.auto_total_dos(plot=False)
                phonopy_normal.run_thermal_properties(temperatures=temp_array)               
                if strain == '1_0':
                    print(phonon_path)
                    # Double for normal and posdef
                    phonon_paths.append(phonon_path)
                    phonon_paths.append(phonon_path)
                    force_constants_normal.append(get_fc_from_pos_dyn_matrices(phonopy_normal))
                    force_constants_posdef.append(get_fc_from_pos_dyn_matrices(phonopy_posdef))
                    
                    print('NORMAL')
                    phonopy_normal.plot_band_structure_and_dos()
                    plt.show()
                    
                therm_prop = phonopy_normal.get_thermal_properties_dict()['free_energy']*0.01036410 #kjmol to ev
                Fvib_per_atom = therm_prop / at_nu_prim
                surface_array_normal[c] = Fvib_per_atom + U_per_atom
                vol_array_normal[c] = vol_per_atom    
             
                
                phonopy_posdef.auto_band_structure(plot=False) 
                phonopy_posdef.auto_total_dos(plot=False)
                phonopy_posdef.run_thermal_properties(temperatures=temp_array)
                if strain == '1_0':
                    print(phonon_path)
                    print('POSDEF')
                    phonopy_posdef.plot_band_structure_and_dos()
                    plt.show()                
                               
                therm_prop = phonopy_posdef.get_thermal_properties_dict()['free_energy']*0.01036410
                Fvib_per_atom = therm_prop / at_nu_prim
                surface_array_posdef[c] = Fvib_per_atom + U_per_atom
                vol_array_posdef[c] = vol_per_atom
                   

            vol_arrays_qha.append(vol_array_normal)
            vol_arrays_qha.append(vol_array_posdef)
            surface_arrays_qha.append(surface_array_normal)
            surface_arrays_qha.append(surface_array_posdef)
```

```python
# Birch murnaghan fit @ each temperature
from mylib import get_lowestE_with_volume2
from mylib import check_poly_fits_given_temp2

# 4th degree polynomial fit @ each temperature
from mylib import get_lowestE_with_volume
from mylib import check_poly_fits_given_temp

# Allows us to do fits in loop (for each member of selected groups).
from mylib import get_freeEvsTemperatures

free_en_name = []
free_en_arrs = []
for c,path in enumerate(phonon_paths):
    if c % 2 ==0:
        free_en_name.append(path.replace('Database1','').replace('1_0/','') + 'Normal')
    if c % 2 ==1:
        free_en_name.append(path.replace('Database1','').replace('1_0/','') + 'Posdef')
        
    a=get_lowestE_with_volume2(temp_array, vol_arrays_qha[c], surface_arrays_qha[c], residuals=False)
    free_en_arrs.append(a[1])
free_en_name
```

```python
def plot_en_diff(index_normal, index_posdef):
    freeEdiff_qha_normal = (free_en_arrs[index_normal[0]] - free_en_arrs[index_normal[1]])*1000
    freeEdiff_qha_posdef = (free_en_arrs[index_posdef[0]] - free_en_arrs[index_posdef[1]])*1000

    print(free_en_name[index_normal[0]])
    print(free_en_name[index_normal[1]])
    print(free_en_name[index_posdef[0]])
    print(free_en_name[index_posdef[1]])

    plt.figure()
    plt.title('QHA free energy difference')
    plt.plot(temp_array, freeEdiff_qha_normal, label='QHA free en difference Normal', c='k' )
    plt.plot(temp_array, freeEdiff_qha_posdef, label='QHA free en difference PosDef', c='r' )
    plt.hlines(0.0, temp_array[0],temp_array[-1], linestyles='dashed')
    # plt.scatter(716,0, c='r', label = 'Transition temp.')
    plt.legend()
    plt.ylabel('meV/atom')
    plt.xlabel('K')
    plt.show()
```
